IMPORTING REQUIRED LIBRARIES

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,Conv2D,Flatten, MaxPooling2D
```

LOADING MNIST HANDWRITTEN DIGIT DATASET

```python
(x_train, y_train),(x_test , y_test)= tf.keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

PRE-VIEWING DATA

```python
fig, axs = plt.subplots(4,5, figsize= (15,15))
plt.gray()
for i, a in enumerate(axs.flat):
    a.matshow(x_train[i])
    a.axis('off')
    a.set_title('Number{}'.format(y_train[i]))
fig.show()    
```

RESHAPING TRAIN & TEST DATASET

```python
input_shape = (28,28,1)
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28 ,1)
```

CREATING SEQUENTIAL MODEL AND ADDING LAYERS

```python
model= Sequential()
model.add(Conv2D(28, kernel_size=(3,3),input_shape= input_shape))
model.add(MaxPooling2D(pool_size= (2,2)))
model.add(Flatten())
model.add(Dense(128, activation = tf.nn.relu)) 
model.add(Dropout(0.2))
model.add(Dense(10, activation = tf.nn.softmax))          
```

COMPILING AND FITTING OF MODEL

```python
model.compile(optimizer = 'adam', metrics=['accuracy'], loss='sparse_categorical_crossentropy')
model.fit(x=x_train, y=y_train , epochs=30)
```

EVALUATION OF DATASET USING TEST DATASET

```python
model.evaluate(x_test,y_test)
```

# ACCURACY = 97.48%
